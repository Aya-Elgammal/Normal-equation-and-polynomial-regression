```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/sample_data/Position_Salaries (1).csv')
data
```

```python
X=data.iloc[:,1:2].values
y=data.iloc[:,2].values
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
linre=LinearRegression()
linre.fit(X,y)  #train
y_pred=linre.predict(X) #test
```

```python
plt.scatter(X,y,color='r')
plt.plot(X,y_pred,color='b')
plt.title('Linear Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=3)
X_poly=poly_reg.fit_transform(X)
linre2=LinearRegression()
linre2.fit(X_poly,y)
X_poly_pred=linre2.predict(X_poly)
```

```python
X_poly
```

```python
X_poly_pred
```

```python
plt.scatter(X,y,color='r')
plt.plot(X,X_poly_pred,color='b')
plt.title('polynomial regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
X_grid=np.arange(min(X),max(X),.1)
X_grid
```

```python
X_grid=X_grid.reshape(len(X_grid),1)
X_grid
```

```python
plt.scatter(X,y,color='r')
plt.plot(X_grid,linre2.predict(poly_reg.fit_transform(X_grid)),color='b')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
